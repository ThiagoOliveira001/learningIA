```python
from bs4 import BeautifulSoup
import re
import unidecode
import pandas as pd
from lxml import html
```

### Faz a leitura dos dados da página html de resultados da caixa e gera um csv

[Página onde se encontra o link para os resultados da lotofacil](https://loterias.caixa.gov.br/Paginas/Lotofacil.aspx)
Acessar o link no final da página: [Resultados da Lotofácil por ordem crescente](https://loterias.caixa.gov.br/Paginas/Download-Resultados.aspx)

Efetue download da página e substitua "Download de Resultados - Portal Loterias_CAIXA.html" caso queira gerar um novo csv

```python
with open('Download de Resultados - Portal Loterias_CAIXA.html', encoding='utf-8') as f:
    page = f.read()
```

```python
soup = BeautifulSoup(page, 'html.parser')
table_results = soup.find(class_='tabela-resultado')
```

```python
def rename_header(text):
    return unidecode.unidecode(re.sub(r'\s+', '_', text.strip()))
```

```python
columns = []
for header in table_results.thead.tr:
    columns.append(rename_header(header.text))
```

```python
df = pd.read_html(str(table_results))[0]
df.info()
```

```python
df[df['Concurso'].astype(str).str.isdigit()].info()
```

```python
df.drop(index=df[~df['Concurso'].astype(str).str.isdigit()].index, axis=0, inplace=True)
```

```python
df.head(5).transpose()
```

```python
df['Concurso'] = df['Concurso'].astype(int)
```

```python
df.rename(columns={
    'Arrecadação_Total': 'Arrecadacao_Total',
    'Ganhadores_15_Números': 'Ganhadores_15_Numeros',
    'Ganhadores_14_Números': 'Ganhadores_14_Numeros',
    'Ganhadores_13_Números': 'Ganhadores_13_Numeros',
    'Ganhadores_12_Números': 'Ganhadores_12_Numeros',
    'Ganhadores_11_Números': 'Ganhadores_11_Numeros',
    'Valor_Rateio_15_Números': 'Valor_Rateio_15_Numeros',
    'Valor_Rateio_14_Números': 'Valor_Rateio_14_Numeros',
    'Valor_Rateio_13_Números': 'Valor_Rateio_13_Numeros',
    'Valor_Rateio_12_Números': 'Valor_Rateio_12_Numeros',
    'Valor_Rateio_11_Números': 'Valor_Rateio_11_Numeros',
    'Acumulado_15_Números': 'Acumulado_15_Numeros',
    'Estimativa_Prêmio': 'Estimativa_Premio',
}, inplace=True)
```

```python
df['Data_Sorteio'] = pd.to_datetime(df['Data_Sorteio'], format='%d/%m/%Y')
```

```python
def convert_money_to_float(value):
    value_without_dot = re.sub(r'[.]+', '', str(value))
    value_with_dot = re.sub(r',', '.', value_without_dot)
    number_value = re.sub(r'[^\d.]+', '', value_with_dot)
    
    return float(number_value)
```

```python
df['Concurso'] = df['Concurso'].astype(int) 

for i in range(1,16):
    df[f'Bola{i}'] = df[f'Bola{i}'].astype(int)

df['Arrecadacao_Total'] = df['Arrecadacao_Total'].apply(lambda x: convert_money_to_float(x))
df['Ganhadores_15_Numeros'] = df['Ganhadores_15_Numeros'].astype(int)

for i in range(14,10, -1):
    df[f'Ganhadores_{i}_Numeros'] = df[f'Ganhadores_{i}_Numeros'].astype(int)
    
for i in range(15,10, -1):
    df[f'Valor_Rateio_{i}_Numeros'] = df[f'Valor_Rateio_{i}_Numeros'].apply(lambda x: convert_money_to_float(x))
    
df['Acumulado_15_Numeros'] = df['Acumulado_15_Numeros'].apply(lambda x: convert_money_to_float(x))
df['Estimativa_Premio'] = df['Estimativa_Premio'].apply(lambda x: convert_money_to_float(x))
df['Valor_Acumulado_Especial'] = df['Valor_Acumulado_Especial'].apply(lambda x: convert_money_to_float(x))
```

```python
df.info()
```

```python
df.to_csv('sorteios_atual.csv',index=False)
```
